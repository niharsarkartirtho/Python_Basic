# 1. Introduction to tupple
#### Tuples are used to store multiple items in a single variable.
#### Tuple is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Set, and Dictionary, all with different qualities and usage.
#### A tuple is a collection which is ordered and unchangeable.
#### Tuples are written with round brackets.
### tupples are Ordered, Unchangeable(list changeable), allow duplicates

```python
thistuple = ("apple", "banana", "cherry")
print(len(thistuple))

#To create a tuple with only one item, you have to add a comma after the item,
#  otherwise Python will not recognize it as a tuple.
thistuple = ("apple",)
print(type(thistuple))
#NOT a tuple
thistuple = ("apple")
print(type(thistuple))

tuple1 = ("apple", "banana", "cherry")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)
tuple4 = ("abc", 34, True, 40, "male")

#tupple constructor
tuple5 = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(tuple5)
```

# 2. Access tuple

```python
tuple1 = ("apple", "banana", "cherry")
print(tuple1[1]) # index start at 0

tuple2 = ("apple", "banana", "cherry")
print(tuple2[-1]) #last index

tuple3 = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(tuple3[2:5]) # idx 2 included 5 excluded

tuple4 = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(tuple4[:4])  # its like [0:4]

tuple5 = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(tuple5[2:]) #cherry to end

tuple6 = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(tuple6[-4:-1]) 

# in keyword with if statement
tuple7 = ("apple", "banana", "cherry")
if "apple" in tuple7:
  print("Yes, 'apple' is in the fruits tuple")
```

# 3. Update tuples

```python
# Change Tuple Values
#Once a tuple is created, you cannot change its values. Tuples are unchangeable, or immutable as it also is called.
#But there is a workaround. You can convert the tuple into a list, change the list, and convert the list back into a tuple.
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

```python
# Add Items
#Since tuples are immutable, they do not have a built-in append() method, 
# but there are other ways to add items to a tuple.

#   1. Convert into a list: Just like the workaround for changing a tuple, 
# you can convert it into a list, add your item(s), and convert it back into a tuple.
tuple1 = ("apple", "banana", "cherry")
y = list(tuple1)
y.append("orange")
tuple1 = tuple(y)
print(tuple1)

#  2. Add tuple to a tuple. You are allowed to add tuples to tuples,
# so if you want to add one item, (or many), create a new tuple with the item(s), 
# and add it to the existing tuple:
tuple2 = ("apple", "banana", "cherry")
y = ("orange",)  #dont forget to add comma
tuple2 += y

print(tuple2)
```

```python
# Remove items
tuple2 = ("apple", "banana", "cherry")
y = list(tuple2)
y.remove("apple")
tuple2 = tuple(y)
print(tuple2)

# delete tuple permenanatly
del tuple2
#print(tuple2) #error because no tuple found
```

# 4. Unpack tuples

```python
# When we create a tuple, we normally assign values to it. This is called "packing" a tuple:
fruits = ("apple", "banana", "cherry")
# But, in Python, 
# we are also allowed to extract the values back into variables. 
# This is called "unpacking":

(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

```python
#Using Asterisk*
# If the number of variables is less than the number of values, 
# you can add an * to the variable name and the values will be assigned to the variable as a list:
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)

#If the asterisk is added to another variable name than the last,
#  Python will assign values to the variable until the number of values left matches the number of variables left.
(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

# 5. loop in tuples

```python
tuple1 = ("apple", "banana", "cherry")
for x in tuple1:
  print(x)

# using index and len()
for i in range(len(tuple1)):
  print(tuple1[i])

# using while loop
i = 0
while i < len(tuple1):
  print(tuple1[i])
  i = i + 1
```

# 6. join tuples  and some methods

```python
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

#using + operator for two of more tuples
tuple3 = tuple1 + tuple2
print(tuple3)

# using * multiply operator for same tuple
tuple3 = tuple2 * 2
print(tuple3)
```

```python
### some extra tuple methods
#### count()	Returns the number of times a specified value occurs in a tuple
#### index()	Searches the tuple for a specified value and returns the position of where it was found
tuple1 = ("apple", "banana", "cherry", "cherry")
print(tuple1.count("cherry"))
print(tuple1.index("cherry"))
```
