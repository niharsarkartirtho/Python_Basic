###                                                                                         4. Variables

```python
#no command/variable type nedded for declaring variables
x = 5
y = "jhon"
print(x) ; print(y)

# no type needded and can change type
a = 6
a = "tirtho"
print(a)

#If you want to specify the data type of a variable, this can be done with casting.
b = str(3)
c = int(4)
d = float(5)

# get the type
print(type(b)) ; print(type(c)) ; print(type(d))

# single qoute or double qoute doensnt matter
print("exit1")
print('exit2')
```

```python
# Many values to Multiple Variables in one line
x , y , z = "Orange" , "Banana" , "Berry"
print(x) ; print(y) ; print(z)

# One Value to Multiple Variables
x = y = z = "Oranges"
print(x) ; print(y) ; print(z)

# Unpack Collections like list/tuple
fruits = ["apple" , "watermelon" , "pineapple" , 2]
x , y , z , a = fruits
print(x) ; print(y) ; print(z) ; print(a)
```

```python
# output multiple variables at a time
x = "python"
y = "is "
z = "awesome"
print(x , y , z)

# String concat operator
print(x + y + z)

# Mathmatical print
a , b = 1 , 2
print(a + b)

# string and numbers cannt be print using +
# print(x + a) error
print(x,a,a)
print(x + str(a))
```

### 5. Datatypes

```python
# Datatypes
x = "Hello World"	#str	
x = 20	#int	
x = 20.5	#float	
x = 1j	#complex	
x = ["apple", "banana", "cherry"]	#list	
x = ("apple", "banana", "cherry")	#tuple	
x = range(6)	#range	
x = {"name" : "John", "age" : 36}	#dict	
x = {"apple", "banana", "cherry"}	#set	
x = frozenset({"apple", "banana", "cherry"})	#frozenset	
x = True	#bool	
x = b"Hello"	#bytes	
x = bytearray(5)	#bytearray	
x = memoryview(bytes(5))	#memoryview	
x = None	#NoneType
print(x)
```

### 6. Strings

```python
# Multiline String
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)

# string is also an array
#python doesnt have char data types
a = "Hello, World!"
print(a[1])

#looping in string
for x in "banana":
  print(x)
  print(len(a))

#Check if "free" is present in the following text we use in keyword
txt = "The best things in life are free!"
print("free" in txt)

## in keyword in an if statement
if "free" in txt:
  print("Yes, 'free' is present.")
  
# not in keyword
print("expensive" not in txt)
```

```python
# 6.1 Slicing string 

# Get the characters from position 2 to position 5 (not included):
b = "Hello, World!"
print(b[2:5])
#Get the characters from the start to position 5 (not included):
print(b[:5])
#to end
print(b[2:])
#
print(b[-5:-2])
```

```python
# 6.2 Modify strings
a = "Hello, World!"
print(a.upper())
print(a.lower())
#The strip() method removes any whitespace from the beginning or the end:

a = " Hello, World! "
print(a.strip()) # returns "Hello, World!" removes any whitespace from the beginning or the end:
print(a.replace("H", "J"))

#The split() method splits the string into substrings if it finds instances of the separator:

a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!']
```
