# 1. Access , Change , Add list items

```python
list1 = ["apple", "banana", "cherry", "apple", "cherry"]
print(list1)
print(len(list1))

#A list with strings, integers and boolean values:
list2 = ["abc", 34, True, 40, "male"]
for x in list1:
    print(x,end=" ")

#It is also possible to use the list() constructor when creating a new list.
list1 = list(("apple", "banana", "cherry")) # note the double round-brackets
print(list1)
```

```python
# Access List items
print(list1[0])
"""Negative Indexing
Negative indexing means start from the end
-1 refers to the last item, -2 refers to the second last item etc.
"""
print(list1[-1]) ; print(list1[-2])

#Range of indexes
print(list2[2:5]) # third , fourth , fifth

#This example returns the items from the beginning to, but NOT including, "kiwi":
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])

#
print(thislist[2:]) #cherry to the end
print(thislist[-4:-1]) # not including mango

```

```python
# Chnage List Items
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)

#Change the values "banana" and "cherry" with the values "blackcurrant" and "watermelon":
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)

#Change the second value by replacing it with two new values:
thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant", "watermelon"]
print(thislist)

#Change the second and third value by replacing it with one value:
thislist = ["apple", "banana", "cherry"]
thislist[1:3] = ["watermelon"]
print(thislist)

# insert method (insert in specefic index)
thislist.insert(3, "watermelon")
print(thislist)

#append() append at the end
thislist.append("orange")

# extend or merge two list
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)

# you can also extend tupple in list
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(thislist)

```

# 2.Loop lists

```python
# loop list
thislist = ['a' , 5 , True , 'banana']
for x in thislist:
    print(x ,end=',')

#using for range loop
print()
for i in range(len(thislist)) :
    print(thislist[i],end=',')

# while loop
print()
i = 0
while i < len(thislist) :
    print(thislist[i])
    i += 1

# short hand
thislist = ["apple", "banana", "cherry"]
[print(x) for x in thislist]
```

# 3. Remove list items

```python
fruits = ["apple" , "banana" , "cherry" , "cherry"]

fruits.remove("apple") #removes apple 
print(fruits)

fruits.append("banana")
fruits.remove("banana") #if there same values removes the first occurence
print(fruits)

fruits[3:3] = ["watermalon" , "malon"]
print(fruits)

fruits.pop(0) #removes specified index
print(fruits)

fruits.pop() #removes last index item
print(fruits)

fruits.append("tomato")
del fruits[0]
print(fruits)

fruits.clear() #list remain but no items
print(fruits)

del fruits
# print(fruits) empty list cannt be printed error

```

## 4. list comprehension

```python
'''
List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.
Example:
Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.
Without list comprehension you will have to write a for statement with a conditional test inside:

'''
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)
print(newlist)

#With list comprehension you can do all that with only one line of code:
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)

#with no if statement
newlist = [x for x in fruits]

newlist = [x for x in range(10)]
print(newlist)

newlist = [x for x in range(10) if x < 5]
print(newlist)

newlist = [x if x != "banana" else "orange" for x in fruits]
print(newlist)
```

## 5.copy list

```python
#You cannot copy a list simply by typing list2 = list1, because:
# list2 will only be a reference to list1, and changes made in list1 will automatically also be made in list2.
# copy() method
fruits= ["apple", "banana", "cherry"]
fruits2 = fruits.copy()
print(fruits2)

# list() method
thislist = ["apple", "banana", "cherry"]
mylist = list(thislist)
print(mylist)

#using slicing
fruit3 = ["apple", "banana", "cherry"]
fruit4 = fruit3[:]
print(fruit4)
```

## 6. list sorting

```python
#sort string by alphabatic(first letter), if uppercase is there it is sorted first
list1 = ["orange", "mango", "kiwi", "pineapple", "banana"]
list1.sort()
print(list1)

list2 = [100, 50, 65, 82, 23]
list2.sort()
print(list2)

#sorting reverse
list3 = ["orange", "mango", "kiwi", "pineapple", "banana"]
list3.sort(reverse = True)
print(list3)

#custom sorting function
def myfunc(n):
  return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)

#built-in key fuction
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)

#only reverse the list
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

## 7. join list

```python
# here are several ways to join, or concatenate, two or more lists in Python.
#One of the easiest ways are by using the + operator.

list1 = ["a", "b", "c"]
list2 = [1, 2, 3]
#using + operator
list3 = list1 + list2
print(list3)

#using for loop and same list
for x in list2:
    list1.append(x)
print(list1)

#using extend method
list1.extend(list2)
print(list1)
```

# Some important list methods
### append()	Adds an element at the end of the list
### clear()	Removes all the elements from the list
### copy()	Returns a copy of the list
### count()	Returns the number of elements with the specified value
### extend()	Add the elements of a list (or any iterable), to the end of the current list
### index()	Returns the index of the first element with the specified value
### insert()	Adds an element at the specified position
### pop()	Removes the element at the specified position
### remove()	Removes the item with the specified value
### reverse()	Reverses the order of the list
### sort()	Sorts the list
