# Dictionaries
### Dictionaries are used to store data values in key:value pairs.
### A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

```python
dict1 = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(dict1)
print(dict1["brand"])

#String, int, boolean, and list data types:
dict2 = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
print(dict2["colors"][0])

#Using the dict() method to make a dictionary:
dict3 = dict(name = "John", age = 36, country = "Norway")
print(dict3)
```

## 2. Access Dictionary items
### dict1["  "] , get() , keys() , values() , items()
### Make a change in the original dictionary, and see that the values , items, keys list gets updated as well even the it is decleared before

```python
#You can access the items of a dictionary by referring to its key name, inside square brackets:
x = dict1["model"]
x = dict1.get("model")
#the keys() method will return a list of all the keys in the dictionary.

x = dict1.keys()
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}
x = car.keys() #automatically change when dic is changed later
print(x) #before the change
car["color"] = "white"
print(x) #after the change  

x = dict2.values()
x = car.values()
print(x) #before the change
car["year"] = 2020
print(x) #after the change

#The items() method will return each item in a dictionary, as tuples in a list.
x = dict2.items()  # key value pair
print(x)

#check if key exits or not
if "model" in dict1:
  print("Yes, 'model' is one of the keys in the thisdict dictionary")
```

## 3. Change values or add items

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["year"] = 2018
print(thisdict)

thisdict.update({"year": 2020})
print(thisdict)

# add items
thisdict["color"] = "red"
#print(thisdict)

#The update() method will update the dictionary with the items from a given argument.
#  If the item does not exist, the item will be added.
thisdict.update({"color": "red"})
print(thisdict)
```

# 4. Remove items 
### pop() , popitem() , del , clear()

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
#pop()
car.pop("model")
print(car)

#popitem() will remove last
car.popitem()
print(car)

#del
del car["brand"]
print(car)

#empty dict
car.clear()
print(car)
```

# 5.Looping through dictionaries

```python
#Print all key names in the dictionary, one by one:
for x in dict1:
  print(x)

print("-----")

#Print all values in the dictionary, one by one:
for x in dict1:
  print(dict1[x])

print("-----")

#You can also use the values() method to return values of a dictionary:
for x in dict1.values():
  print(x)

print("-----")

#You can use the keys() method to return the keys of a dictionary:
for x in dict1.keys():
  print(x)

print("-----")

#Loop through both keys and values, by using the items() method:
for x, y in dict1.items():
  print(x, y)
```

# 6. copy dictionaries
#### You cannot copy a dictionary simply by typing dict2 = dict1, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

```python
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
# copy()
car2 = car.copy()
print(car2)

# dict()
car3 = dict(car2)
print(car3)
```

# 7. Nested dictionaries
#### A dictionary can contain dictionaries, this is called nested dictionaries.

```python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}

#Create three dictionaries, then create one dictionary that will contain the other three dictionaries:
child1 = {
  "name" : "Emil",
  "year" : 2004
}
child2 = {
  "name" : "Tobias",
  "year" : 2007
}
child3 = {
  "name" : "Linus",
  "year" : 2011
}

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}

print(myfamily["child2"]["name"])
```

```python
#You can loop through a dictionary by using the items() method like this:
#Loop through the keys and values of all nested dictionaries:

for x, obj in myfamily.items():
  print(x)
  for y in obj:
    print(y + ':', obj[y])
```
